```python
def cal_ratio(dataframe):
    fraud_tran = dataframe[dataframe['isFraud'] == 1]
    normal_tran = dataframe[dataframe['isFraud'] == 0]
    print(f"The number of fraud: {len(fraud_tran)}")
    print(f"The number of fraud: {len(normal_tran)}")
    print(f"The ratio is {len(fraud_tran)/len(normal_tran)}")
cal_ratio(train_tranc_df)
```
